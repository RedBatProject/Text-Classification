```python
# vocab_size, nmbed, nmbed * block_size,
nhidden = 1000
block_size = 10
```

```python
import torch.nn as nn
import torch.nn.functional as F

# PyTorch models inherit from torch.nn.Module
class GarmentClassifier(nn.Module):
    def __init__(self):
        super(GarmentClassifier, self).__init__()
        self.emb = Embedding(vocab_size, nmbed)
        self.flat = Flatten()
        self.linear1 = Linear(nmbed * block_size, nhidden)
        self.linear2 = Linear(nhidden, 2*nhidden)
        self.linear3 = Linear(2*nhidden, nhidden)

        self.batchnorm = BatchNorm1d(nhidden)
        # self.tan = Tanh()
        self.out = Linear(nhidden, vocab_size)

    def forward(self, x):
        x = self.emb(x)
        x = self.flat(x)
        x = F.relu(self.linear1(x))
        x = F.relu(self.linear2(x))
        x = F.relu(self.linear3(x))

        x = self.batchnorm(x)
        x = F.relu(x)
        x = self.out(x)
        return x


model = GarmentClassifier()
```

```python
model = model.to('cuda')
```
